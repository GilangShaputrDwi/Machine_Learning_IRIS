2. ubahlah nilai suhu_C menjadi nilai suhu dalam fahrenheit
3. - Jika suhu Fahrenheit diatas 150, maka cetak â€œSekarang suhunya panasâ€
  - Jika suhu Fahrenheit diatas 100, maka cetak â€œSekarang suhunya normalâ€
  - Jika suhu Fahrenheit selain nilai diatas, maka cetak â€œSekarang suhunya dinginâ€

````python
# Meminta input dari pengguna
suhu_C = float(input("Masukkan suhu dalam Celcius: "))

# Konversi ke Fahrenheit
suhu_F = (suhu_C * 9/5) + 32

# Menampilkan hasil
print(f"Suhu dalam Fahrenheit: {suhu_F:.2f}Â°F")
````

````python
# Meminta input dari pengguna
suhu_F = float(input("Masukkan suhu dalam Fahrenheit: "))

# Konversi ke Celcius
suhu_C = (suhu_F - 32) * 5/9

# Menampilkan hasil
print(f"Suhu dalam Celcius: {suhu_C:.2f}Â°C")
````

````python
# Meminta input dari pengguna
suhu_F = float(input("Masukkan suhu dalam Fahrenheit: "))

# Konversi ke Celcius
suhu_C = (suhu_F - 32) * 5/9

# Menampilkan hasil konversi
print(f"Suhu dalam Celcius: {suhu_C:.2f}Â°C")

# Menentukan kategori suhu
if suhu_F > 150:
    print("Sekarang suhunya panas")
elif suhu_F > 100:
    print("Sekarang suhunya normal")
else:
    print("Sekarang suhunya dingin")
````
